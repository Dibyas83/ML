```python


df5 = pd.DataFrame({
    'FirstName': ['John', 'Jane'],
    'LastName': ['Smith', 'Doe'],
    'City': ['New York', 'Los Angeles']
})

# Combine FirstName and LastName into a 'FullName' column with a space separator
df5['FullName'] = df5['FirstName'] + ' ' + df5['LastName']

# Combine all three columns into an 'Address' column with comma and space separators
df5['Address'] = df5['FullName'] + ', ' + df5['City']

print(df5)
```

```python

df6 = pd.DataFrame({
    'First': ['Ganesh', 'Sartaj'],
    'Last': ['Gaitonde', 'Singh']
})

# Combine 'First' and 'Last' into 'Name' with a space separator
df6['Name'] = df6['First'].str.cat(df6['Last'], sep=' ')

print(df6)
```

```python
df7 = pd.DataFrame({
    'Salary': [50000, 60000],
    'Bonus': [5000, 7000]
})

# Calculate 'TotalCompensation' by adding 'Salary' and 'Bonus'
df7['TotalCompensation'] = df7['Salary'] + df7['Bonus']

print(df7)
```

```python
df8 = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-02'],
    'Time': ['10:00:00', '11:30:00'],
    'Event': ['Meeting', 'Presentation']
})

# Combine 'Date' and 'Time' into a 'DateTime' column and convert to datetime objects
df8['DateTime'] = df8.apply(lambda row: pd.to_datetime(row['Date'] + ' ' + row['Time']), axis=1)

# Combine 'Event' and 'DateTime' into a 'Log' string
df8['Log'] = df8.apply(lambda row: f"{row['DateTime']} - {row['Event']}", axis=1)

print(df8)
```

Converter - for transformation over cols

```python
pd.read_csv('Sold Player.csv',encoding='ansi')
```

```python
def rename(name):
    if name == "Sunrisers Hyderabad":
        return 'SH'
    else:
        return name
```

```python
rename("Sunrisers Hyderabad")
```

pd.read_csv('Sold Player.csv',encoding='ansi',converters={'Teams':rename})
